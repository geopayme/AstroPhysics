## Step 8: Indirectness Scoring
We assess 'indirectness' by comparing results across subgroups (e.g., workplace type). Large differences may indicate indirectness.

```python

# Boxplot comparison across Workplace
sns.boxplot(x='Workplace', y='CCF', data=df)
plt.title('Climate Crisis Factor (CCF) Across Workplace Types')
plt.show()

# Estimate subgroup means
group_means = df.groupby('Workplace')['CCF'].mean()
print("Mean CCF by Workplace:")
print(group_means)

# Calculate percent deviation from mean
overall_mean = df['CCF'].mean()
indirectness_scores = (group_means - overall_mean).abs() / overall_mean * 100
print("\n% Indirectness from group means:")
print(indirectness_scores)
```
